```python
import pandas as pd #using pandas to work with the data set
import seaborn as sns #using seaborn to plot a correlational matrix as a heatmap
import matplotlib.pyplot as plt #using matplotlib to visualize the dataset as graphs and histograms
```

```python
data = pd.read_csv('RedWineQuality.csv') #reading the csv file and storing it in the variable data.
data.head() #showing the first five values from the data sheet.
```

```python
  data.describe() #describing the data using different standards
```

```python
newdata = data.dropna() #using function "dropna" in order to eliminate rows containing NULL Values
newdata.head()
```

```python
data.hist(column='quality') #using .hist function to create the histogram of the coloumn quality
plt.title('Histogram of Quality Score') #Title of the histogram
plt.xlabel('Quality Score') #Contents of the X axis
plt.ylabel('Frequency') #Contents of the Y axis
plt.show() #displaying the histogram
```

```python
# From the Histogram we could conclude that... most of the Red wine have the quality score of 5.
```

```python
X=data['alcohol'] #storing all the values in the coloumn quality in the variable X
Y=data['quality'] #storing all the values in the coloumn alcohol in the variable Y
```

```python
plt.figure(figsize=(10,6)) #determining the size of the graph
plt.scatter(X,Y, label='Quality score',color='red',s=2) #using .scatter function to scatter plot the data labled as Quality score, with color red and size of dots as 2.
plt.xlabel('Alcohol') #Labeling X as Quality
plt.ylabel('Quality') #Labeling Y as Alcohol Content
plt.title('Scatter plot to visualise the relationship between alcohol and quality') #Title of the graph
plt.legend()
X=X.fillna(0) #Replacing the NULL Values with 0
Y=Y.fillna(0)
plt.show() #Displaying the graph
```

```python
# From the scatter graph we could conclude that
# quality 3 alcohol have generally lesser alcohol content and quality 8 alcohol have generally higher alcohol content
```

```python
plt.figure(figsize=(16, 6)) #settingup the size of the figure
heatmap = sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG') #using .heatmap function to create a heatmap of the data
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12) #setting up the title of the heatmap and determining the fontsize.
```

```python
# From the correlational heatmap we could understand the relationship and indipendency of different factors.
```
